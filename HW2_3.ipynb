## Loading NetCDF files

```python
tser = 0
prof = 0
moms = 0


def read_moments(nr):
    # Using readlines()
    file1 = open('moments.00'+str(nr)+'.txt', 'r')
    Lines = file1.readlines()
    num_t = 33
    count = 0
    j = 0
    sfs_tke = []
    uu = []
    vv = []
    ww = []
    t_start = 11
    # Strips the newline character
    for line in Lines:
        count += 1
        if ((count - 18) % 135) == 0:
    #         print("Line{}: {}".format(count, line.strip()))
            sfs_tke.append(float(Lines[count].split()[12]))
            uu.append(float(Lines[count].split()[7]))
            vv.append(float(Lines[count].split()[8]))
            ww.append(float(Lines[count].split()[10]))
            j += 1

    uu = np.array(uu)[t_start:]
    vv = np.array(vv)[t_start:]
    ww = np.array(ww)[t_start:]
    sfs_tke = np.array(sfs_tke)[t_start:]
    t = np.arange(num_t)[t_start:] * 10 + 10
    res_tke = (uu + vv + ww) / 2 
    plt.figure()
    plt.plot(t, sfs_tke, label='subgrid TKE')
    plt.plot(t, res_tke, label='resolved TKE')
    plt.plot(t, sfs_tke + res_tke, label='total TKE')
    plt.legend(fontsize=14)
    plt.xlabel('t(min)', fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.ylabel(r'TKE ($m^2/s^2$)', fontsize=14)
    plt.title(r'Simulation with $\Delta x=\Delta y=$' + str(np.round(3.125* 2**(nr), 4))+'m')
    plt.show()

read_moments(3)
read_moments(2)
```
